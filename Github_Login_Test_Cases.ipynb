```python
import json
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
import warnings
warnings.filterwarnings('ignore')
```

```python
credentials = open("input.json")
data = json.load(credentials)
correct_email = data['correct email']
correct_password = data['correct password']
incorrect_email = data['incorrect email']
incorrect_password = data['incorrect password']
```

# Test Case 1 : Correct Email & Correct Password

```python
url = "https://github.com/login"
driver = webdriver.Firefox(executable_path='/tmp/geckodriver')
driver.get(url)

driver.find_element_by_id('login_field').send_keys(correct_email)

driver.find_element_by_id('password').send_keys(correct_password)

driver.find_element_by_name('commit').click()

WebDriverWait(driver=driver, timeout=10).until(
    lambda x: x.execute_script("return document.readyState === 'complete'")
)

error_message = "Incorrect username or password."

errors = driver.find_elements_by_class_name("flash-error")

if any(error_message in e.text for e in errors):
    print("Login failed")
    
else:
    print("Login successful")
```
